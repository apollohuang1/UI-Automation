```python
import xmltodict
```

```python
from ppadb.client import Client as AdbClient
client = AdbClient(host="127.0.0.1", port=5037)
```

```python
device = client.devices()[0]
```

```python
device
```

```python
device.shell('uiautomator dump')
```

```python
device.pull('/sdcard/window_dump.xml', 'data/window_dump.xml')
```

```python
dic = xmltodict.parse(open('data/window_dump.xml', 'r').read())
```

```python
import json
```

```python
json.dump(dic, open('window_dump.json', 'w'), indent=4)
```
